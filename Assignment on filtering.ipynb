```python
import pandas as pd
```

```python
df=pd.read_csv('Downloads/Cars93.csv')
```

```python
df.columns
```

```python
#1. How can you filter cars with horsepower greater than 200?
df[['Model','Horsepower']][df['Horsepower']>200]
```

```python
#2. Filter cars with horsepower less than or equal to 100.
df[['Model','Horsepower']][df['Horsepower']<=100]
```

```python
#3. How can you extract cars with exactly 4 cylinders?
df[['Model','Cylinders']][df['Cylinders']=='4']

```

```python
#4. Filter cars with a price higher than $25,000
df[['Model','Price']][df['Price']>25]
```

```python
#5. How can you extract cars with a weight between 3000 and 3500 lbs?
df[['Model','Weight']][(df['Weight']>=3000)&(df['Weight']<=3500)]
```

```python
#6. How can you extract cars with an MPG.city greater than or equal to 25?
df[['Model','MPG.city']][df['MPG.city']>=25]
```

```python
#7. Filter cars with an MPG.highway less than 20.

df[df['MPG.highway']<20]
```

```python
#8. How can you extract cars that have a 'Type' of 'Sporty' or 'Compact'?
df['Type'][(df['Type']=='Sporty')|(df['Type']=='Compact')]
```

```python
#9. Filter cars that are either 'Domestic' or 'Imported'
df['Origin'][(df['Origin']=='USA')|(df['Origin']=='non USA')]
df['Type'][(df['Type']=='Sporty')|(df['Type']=='Compact')]
```

```python
#10. How can you extract cars with NaN values in the 'Rear.seat.room' column?
df[df['Rear.seat.room'].isna()]
```

```python
#11. Filter cars that have a 'Manufacturer' name starting with the letter 'C'.
df[df['Manufacturer'].str.startswith('C')]
```

```python
#12. How can you extract cars with 'AirBags' labeled as 'Non
len(df[df['AirBags']=='None'])
```

```python
#13. Filter cars with 'Man.trans.avail' set to 'Yes'.
df[df['Man.trans.avail']=='Yes']
```

```python
#14. Filter cars that have a 'Turn.circle' less than 35.
df[df['Turn.circle']<35]
```

```python
#15. How can you extract cars that have 'Rev.per.mile' greater than 240?
df[df['Rev.per.mile']>240]
```

```python
#16. Filter cars that have 'MPG.city' and 'MPG.highway' values greater than 20.
df[(df['MPG.city']>20)&(df['MPG.highway']>20)]
```

```python
#17. How can you extract cars with a 'Fuel.tank.capacity' less than or equal to 15 gallons?
df[df['Fuel.tank.capacity']<=15]
```

```python
#18. Filter cars with a 'Length' greater than 190 inches
df[df['Length']>190]
```
